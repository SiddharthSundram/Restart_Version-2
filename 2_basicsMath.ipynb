Problem 4 — Find Largest Digit in a Number

```python
def largestDigit(n):
    if n < 0:
        n = -n
    largest = 0
    while n > 0:
        digit = n % 10
        if digit > largest:
            largest = digit
        n = n // 10
    return largest
n = int(input())
print(largestDigit(n))
```

Problem 3 — Reverse Number

```python
def reverseNum(n):
    temp = n
    if n < 0:
        n = -n
    rev = 0
    while n > 0:
        d = n % 10
        rev = rev * 10 + d
        n = n // 10
    if temp < 0 :
        return -rev
    return rev
n = int(input())
print(reverseNum(n))
```

Problem 4 — Check Palindrome Number

```python
def palindrome(n):
    temp = n
    if n < 0:
        return f'{n} negative number. Not Palindrome'
    rev = 0
    while n > 0:
        d = n % 10
        rev = rev * 10 + d
        n = n // 10
        
    if temp == rev:
        return 'Palindrome'
    else:
        return 'Not Palindrome'
n = int(input())
print(palindrome(n))
```

Problem 1 — Smallest Digit in Number

```python
def smallestDigit(n):
    n = abs(n)
    if n == 0:
        return 0
    smalldigit = 9
    while n > 0:
        d = n % 10
        if d < smalldigit:
            smalldigit = d
        n = n // 10
    return smalldigit
n = int(input())
print(smallestDigit(n))
```

Problem 2 — Count Digits Equal to First Digit (from left)

```python
def countFirstDigit(n):
    i = 0
    countFdigit = 0
    fdigit = 0
    while n > 0:
        d = n % 10
        if i == 0:
            fdigit = d
            i += 1
        if d == fdigit :
            countFdigit += 1
        n = n // 10
    return countFdigit
n = int(input())
print(countFirstDigit(n))
```

Problem 3 — Check if All Digits are Even

```python
def evencheck(n):
    n = abs(n)
    while n > 0:
        d = n % 10
        if d % 2 != 0:
            return False
        n //= 10
    return True
n = int(input())
print(evencheck(n))
```

Problem 4 — Find Difference Between Largest and Smallest Digit

```python
def diff(n):
    n = abs(n)
    large = 0
    small = 9
    while n > 0:
        d = n % 10
        if d > large:
            large = d
        if d < small:
            small = d
        n //= 10
    return large - small
n = int(input())
print(diff(n))
```

Problem 2 — Count Digits Equal to First Digit

```python
def countFirstDigit(n):
    n = abs(n)
    if n == 0:
        return 1
    temp = n
    countFdigit = 0
    fdigit = 0
    while n > 0:
        d = n % 10
        fdigit = d
        n //= 10
    while temp > 0:
        d = temp % 10
        if fdigit == d:
            countFdigit += 1
        temp //= 10
    return countFdigit
n = int(input())
print(countFirstDigit(n))
```
