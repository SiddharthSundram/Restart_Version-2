Majority Element

```python
def majorityElement(nums):
        count = maj = 0 ; checked = [] 
        element = nums[0]
        for i in nums:
            if i not in checked :
                for j in nums:
                    if i == j:
                        count += 1
                if count > maj:
                    maj = count
                    element = i
                count = 0
                checked.append(i)
        if maj > len(nums) // 2:
            return element
        else:
            return -1

arr = list(map(int,input().strip().split()))
print(majorityElement(arr))
```

```python
# boyer -moore algo
def majEl(arr):
    count = 0 ; maj = None
    for i in arr:
        if count == 0 :
            maj = i
            count += 1
        else:
            if i == maj :
                count += 1
            else:
                count -= 1
    count = 0
    for i in arr: 
        if maj == i:
            count += 1
    if count > len(arr) // 2:
        return maj
    else:
        return -1

arr = list(map(int,input().strip().split()))
print(majEl(arr))
```

Right rotate an array k-time in place.

```python
def rotateOnce(arr):
    temp = arr[len(arr)-1]
    for i in range(len(arr)-2,-1,-1):
        arr[i+1] = arr[i]
    arr[0] = temp
    return arr

def rotateArr(arr,k):
    for i in range(k):
        rotateOnce(arr)
    return arr

arr = list(map(int,input().strip().split()))
k = int(input())
print(rotateArr(arr,k))
        
```

```python
def rotateArr(arr,k):
    
    kl = len(arr)-k ; kr = len(arr)-1
    fl = 0 ; fr = len(arr)-1-k
    l = 0 ; r = len(arr) -1
    
    def reverseRotate(l,r):
        while l < r:
                arr[l],arr[r] = arr[r],arr[l]
                l += 1
                r -= 1
        return arr
    reverseRotate(kl,kr)
    reverseRotate(fl,fr)
    reverseRotate(l,r)
    
    return arr
rotateArr([1,2,3,4,5,6,7,8],3)
```

Find the element that occuring odd number of times (only element that occurs odd number of times)

```python
def countOdd(arr):
    count = 0
    for i in arr:
        for j in arr:
            if i == j:
                count += 1
        if count % 2 != 0:
            return i
        count = 0
    return -1
    
countOdd([1,2,3,2,3,1,3])
```

```python
def oddCount(arr):
    myDict = {}
    for i in arr:
        myDict[i] = myDict.get(i,0)+1
    for k,v in myDict.items():
        if v % 2 != 0:
            return k
    return -1
oddCount([1,2,3,2,3,1,3])
```

```python
def countOdd(arr):
    count = 0
    for i in arr:
        count ^= i
    return count
    
countOdd([0,0,1,2,3,2,3,1,0,0,0])
```

How Many Numbers Are Smaller Than the Current Number


```python
def smallerThanCurr(arr):
    result = []
    for i in arr:
        count = 0
        for j in arr:
            if i>j:
                count +=1
        result.append(count)
    return result

smallerThanCurr([8,1,2,2,3])
```

1389. Create Target Array in the Given Order


```python
def createTarget(nums,index):
    target = []
    for i in range(len(nums)):
        if index[i] >= len(nums):
            target.append(nums[i])
        else:
            target.insert(index[i], nums[i])
    return target


nums = list(map(int,input().split()))
index = list(map(int,input().split()))
createTarget(nums,index)
```

1299. Replace Elements with Greatest Element on Right Side

```python
def greatestRight(arr):
    l = len(arr)
    for i in range(l):
        if i == l - 1:
            arr[i] = -1
        else:
            if i < l: 
                max = arr[i+1] 
            for j in range(i+1,l):
                if arr[j] > max:
                    max = arr[j]
            arr[i] = max 
    return arr

# input_arr = [17,18,2,3,5,6,1]
arr = list(map(int,input().split()))
greatestRight(arr)
```
