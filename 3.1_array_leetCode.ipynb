Majority Element

```python
def majorityElement(nums):
        count = maj = 0 ; checked = [] 
        element = nums[0]
        for i in nums:
            if i not in checked :
                for j in nums:
                    if i == j:
                        count += 1
                if count > maj:
                    maj = count
                    element = i
                count = 0
                checked.append(i)
        if maj > len(nums) // 2:
            return element
        else:
            return -1

arr = list(map(int,input().strip().split()))
print(majorityElement(arr))
```

```python
def majorityElement(nums):
        element = {}
        for i in nums:
            element[i] = element.get(i,0) + 1
        
        for key,val in element.items():
            if val > len(nums) // 2:
                return key
        return -1

arr = list(map(int,input().strip().split()))
print(majorityElement(arr))
```

```python
# boyer -moore algo
def majEl(arr):
    count = 0 ; maj = None
    for i in arr:
        if count == 0 :
            maj = i
            count += 1
        else:
            if i == maj :
                count += 1
            else:
                count -= 1
    count = 0
    for i in arr: 
        if maj == i:
            count += 1
    if count > len(arr) // 2:
        return maj
    else:
        return -1

arr = list(map(int,input().strip().split()))
print(majEl(arr))
```

Right rotate an array k-time in place.

```python
def rotateOnce(arr):
    temp = arr[len(arr)-1]
    for i in range(len(arr)-2,-1,-1):
        arr[i+1] = arr[i]
    arr[0] = temp
    return arr

def rotateArr(arr,k):
    for i in range(k):
        rotateOnce(arr)
    return arr

arr = list(map(int,input().strip().split()))
k = int(input())
print(rotateArr(arr,k))
        
```

```python
def rotateArr(arr,k):
    temp = []
    for i in range(len(arr)-k,len(arr)):
        temp.append(arr[i])
    for i in range(len(arr)-1-k,-1,-1):
        arr[i+k] = arr[i]
    for i in range(k):
        arr[i] = temp[i]
    return arr

rotateArr([1,2,3,4,5,6,7],3)
```

```python
def rotateArr(arr,k):
    kl = len(arr)-k ; kr = len(arr)-1
    fl = 0 ; fr = len(arr)-1-k
    l = 0 ; r = len(arr) -1
    for i in range(len(arr)):
        if kl < kr:
            arr[kl],arr[kr] = arr[kr],arr[kl]
            kl += 1
            kr -= 1
        if fl < fr:
            arr[fl],arr[fr] = arr[fr],arr[fl]
            fl += 1
            fr -= 1
    for i in range(len(arr)):
        if l < r:
            arr[l],arr[r] = arr[r],arr[l]
            l += 1
            r -= 1
    return arr
rotateArr([1,2,3,4,5,6,7],3)
```

```python
def rotateArr(arr,k):
    
    kl = len(arr)-k ; kr = len(arr)-1
    fl = 0 ; fr = len(arr)-1-k
    l = 0 ; r = len(arr) -1
    
    def reverseRotate(l,r):
        while l < r:
                arr[l],arr[r] = arr[r],arr[l]
                l += 1
                r -= 1
        return arr
    reverseRotate(kl,kr)
    reverseRotate(fl,fr)
    reverseRotate(l,r)
    
    return arr
rotateArr([1,2,3,4,5,6,7,8],3)
```

Find the element that occuring odd number of times (only element that occurs odd number of times)

```python
def countOdd(arr):
    count = 0
    for i in arr:
        for j in arr:
            if i == j:
                count += 1
        if count % 2 != 0:
            return i
        count = 0
    return -1
    
countOdd([1,2,3,2,3,1,3])
```

```python
def oddCount(arr):
    myDict = {}
    for i in arr:
        myDict[i] = myDict.get(i,0)+1
    for k,v in myDict.items():
        if v % 2 != 0:
            return k
    return -1
oddCount([1,2,3,2,3,1,3])
```

```python
def countOdd(arr):
    count = 0
    for i in arr:
        count ^= i
    return count
    
countOdd([0,0,1,2,3,2,3,1,0,0,0])
```
