Majority Element

```python
def majorityElement(nums):
        count = maj = 0 ; checked = [] 
        element = nums[0]
        for i in nums:
            if i not in checked :
                for j in nums:
                    if i == j:
                        count += 1
                if count > maj:
                    maj = count
                    element = i
                count = 0
                checked.append(i)
        if maj > len(nums) // 2:
            return element
        else:
            return -1

arr = list(map(int,input().strip().split()))
print(majorityElement(arr))
```

```python
def majorityElement(nums):
        element = {}
        for i in nums:
            element[i] = element.get(i,0) + 1
        
        for key,val in element.items():
            if val > len(nums) // 2:
                return key
        return -1

arr = list(map(int,input().strip().split()))
print(majorityElement(arr))
```

```python
# boyer -moore algo
def majEl(arr):
    count = 0 ; maj = None
    for i in arr:
        if count == 0 :
            maj = i
            count += 1
        else:
            if i == maj :
                count += 1
            else:
                count -= 1
    count = 0
    for i in arr: 
        if maj == i:
            count += 1
    if count > len(arr) // 2:
        return maj
    else:
        return -1

arr = list(map(int,input().strip().split()))
print(majEl(arr))
```

Right rotate an array k-time in place.

```python
def rotateOnce(arr):
    temp = arr[len(arr)-1]
    for i in range(len(arr)-2,-1,-1):
        arr[i+1] = arr[i]
    arr[0] = temp
    return arr

def rotateArr(arr,k):
    for i in range(k):
        rotateOnce(arr)
    return arr

arr = list(map(int,input().strip().split()))
k = int(input())
print(rotateArr(arr,k))
        
```
