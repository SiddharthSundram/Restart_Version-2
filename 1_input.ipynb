Input:

2 3
1 2 3
4 5 6


Output:
Print matrix sum.

```python
def mat(arr):
    sum = 0
    for i in arr:
        for j in i:
            sum += j
    return sum

m,n = map(int,input().split())
arr = [list(map(int,input().split()))  for i in range(m)]
print(mat(arr))
```

Reverse Number (Manual)

```python
def revNum(n):
    rev = 0
    while n > 0:
        digit = n % 10
        rev = digit + rev*10
        n = n // 10
    return rev

n = int(input())
print(revNum(n))
```

Problem 2 — Count Digits

```python
def countDigit(n):
    count = 0
    while n > 0:
        count += 1
        n = n // 10
    return count

n = int(input())
print(countDigit(n))
```

Problem 3 — Find Maximum in Array (Manual)

```python
def MaxVal(n):
    maximumVal = n[0]
    for i in n:
        if maximumVal < i:
            maximumVal = i
    return maximumVal

arr = list(map(int,input().split()))
print(MaxVal(arr))
```

Problem 4 — Count Vowels in String

```python
def countVowel(st):
    vowel = 'aeiou'
    count = 0
    for i in st:
        if i in vowel:
            count += 1
    return count

st = input()
print(countVowel(st))
```

Problem 5 — Sum Until User Enters 0

```python
def sumUntilZero(n):
    total = 0
    while n != 0:
        total += n
        n = int(input())
    return total

n = int(input())
print(sumUntilZero(n))
```
