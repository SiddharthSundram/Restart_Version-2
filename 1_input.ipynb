```python
def arith(m,n):
    return (m+n),(m-n)

m,n = map(int,input().split())
add,diff = arith(m,n)
print(f'Sum = {add}, Difference = {diff}')
```

Input:
1 2 3 4 5
Output:
Print array and its sum.

```python
def arr(n):
    total = 0
    for i in n:
        total += i
    return total

n = list(map(int,input().split()))
print('Array =',n,'and Sum of array =', arr(n))
```

Q6

Input:

2 3
1 2 3
4 5 6


Output:
Print matrix sum.

```python
def mat(arr):
    sum = 0
    for i in arr:
        for j in i:
            sum += j
    return sum

m,n = map(int,input().split())
arr = [list(map(int,input().split()))  for i in range(m)]
print(mat(arr))
```
