Problem 1 — Count Positive Numbers in Array

```python
def countPositive(arr):
    if arr == []: return "Empty array"
    countPN = 0
    for i in arr:
        if i > 0:
            countPN += 1
    return countPN
arr = list(map(int,input().strip().split()))
print(countPositive(arr))
```

Problem 2 — Find Largest Number in Array (Manual)

```python
def largestNo(arr):
    if arr == [] : return "Empty Array/List"
    large = arr[0]
    for i in arr:
        if i > large:
            large = i
    return large
arr = list(map(int,input().split()))
print(largestNo(arr))
```

Problem 3 — Sum of Even Numbers in Array

```python
def evenSum(arr):
    if arr == []: return 'empty arr'
    total = 0
    for i in arr:
        if i % 2 == 0: total += i
    return total
arr = list(map(int,input().split()))
print(evenSum(arr))
```

Problem 4 — Count Occurrences of a Given Number

```python
def occuranceCount(arr,target): 
    if arr == [] : return 'Empty array' 
    countTarget = 0 
    for i in arr: 
        if i == target: 
            countTarget += 1 
    return countTarget 
arr = list(map(int,input().split())) 
target = int(input()) 
print(occuranceCount(arr,target)) 
```

Problem 1 — Find Second Largest Number

```python
def secondLargest(arr):
    arr = list(set(arr))
    if len(arr) <= 1: return None
    if arr[0] > arr[1]:
        first = arr[0]
        second = arr[1]
    else:
        first = arr[1]
        second = arr[0]
    
    for i in arr:
        if i > first:
            second = first
            first = i
        elif first > second and i > second:
            second = i
    
    return second
arr = list(map(int,input().split()))
print(secondLargest(arr))
        
```

Problem 2 — Count Elements Greater Than Average

```python
def countGreaterAvgEl(arr):
    if arr == []: return 0
    total = 0
    countAvg = 0
    for i in arr:
        total += i
    avge = total/len(arr)
    for i in arr: 
        if i > avge: countAvg += 1
    return countAvg
arr = list(map(int,input().split()))
print(countGreaterAvgEl(arr))
```

Problem 3 — Reverse Array Manually

```python
def arrRev(arr):
    if len(arr) <= 0: return 'Empty Array'
    rev = []
    for i in range(len(arr)-1,-1,-1):
        # print(arr[i],end=' ')
        rev.append(arr[i])
    return rev
arr = list(map(int,input().split()))
print(arrRev(arr))
# arrRev([1,2,3,4,5,6])
    
```

Problem 4 — Check if Array is Sorted (Ascending)

```python
def sortedArr(arr):
    if arr == [] : return 'Empty array'
    for i in range(len(arr)):
        if i < len(arr)-1:
            if arr[i] > arr[i+1]:
                return False
    return True

arr = list(map(int,input().split()))
print(sortedArr(arr))
```

Problem 1 — Count Increasing Adjacent Pairs

```python
def countIncPair(arr):
    if arr == [] : return 'Empty Array'
    countPair = 0
    for i in range(len(arr)-1):
        if arr[i] < arr[i+1]:
            countPair += 1
    return countPair
arr = list(map(int,input().split()))
print(countIncPair(arr))
```

Problem 2 — Maximum Adjacent Difference

```python
def maxAdjDiff(arr):
    if arr == [] : return 'Empty Array'
    maxDiff = 0 ; currDiff = 0
    for i in range(len(arr)-1):
        currDiff = abs(arr[i+1] - arr[i])
        if maxDiff < currDiff:
            maxDiff = currDiff
    return maxDiff
arr = list(map(int,input().split()))
print(maxAdjDiff(arr))
```

Problem 3 — Check Palindrome Array

```python
def palinCheck(arr):
    if arr == [] : return 'Empty array'
    rev = []
    for i in range(len(arr)-1,-1,-1):
        rev.append(arr[i])
    return rev == arr
arr = list(map(int,input().split()))
print(palinCheck(arr))
```

Problem 4 — Move All Zeros to End (Stable)

```python
def moveZero(arr):
    if arr == [] : return 'Empty arrray'
    for i in arr:
        if i == 0:
            arr.remove(0)
            arr.append(0)
    return arr
arr = list(map(int,input().split()))
print(moveZero(arr))
```
