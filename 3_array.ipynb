Problem 1 — Count Positive Numbers in Array

```python
def countPositive(arr):
    if arr == []: return "Empty array"
    countPN = 0
    for i in arr:
        if i > 0:
            countPN += 1
    return countPN
arr = list(map(int,input().strip().split()))
print(countPositive(arr))
```

Problem 2 — Find Largest Number in Array (Manual)

```python
def largestNo(arr):
    if arr == [] : return "Empty Array/List"
    large = arr[0]
    for i in arr:
        if i > large:
            large = i
    return large
arr = list(map(int,input().split()))
print(largestNo(arr))
```

Problem 3 — Sum of Even Numbers in Array

```python
def evenSum(arr):
    if arr == []: return 'empty arr'
    total = 0
    for i in arr:
        if i % 2 == 0: total += i
    return total
arr = list(map(int,input().split()))
print(evenSum(arr))
```

Problem 4 — Count Occurrences of a Given Number

```python
def occuranceCount(arr,target): 
    if arr == [] : return 'Empty array' 
    countTarget = 0 
    for i in arr: 
        if i == target: 
            countTarget += 1 
    return countTarget 
arr = list(map(int,input().split())) 
target = int(input()) 
print(occuranceCount(arr,target)) 
```
